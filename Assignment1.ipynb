```python
# Write a Python program that takes user input for two numbers and performs all arithmetic operations on them.  
#Display the results with appropriate messages for each operation. 
```

```python
a = int(input("enter your first number:"))
b = int(input("enter your second number:"))


#addition
print("your sum of the two number is", a+b)


# subtraction
print("your difference of the two number is", a-b)



#multiplication
print("your product of two number is", a*b)



#division
print("your quotient of two number is", a/b)


# modulus: to find the remainder
print("your remainder of two number is",a%b)


#floor division: to find the quotient without decimal points
print("your quotient of two number is", a//b) #without decimal



#exponent: raises one number to the power of the other 
print("your exponent value of two number is", a**b)
```

```python
#Write a python program to interchange a list items with & without using temporary values. 
```

```python
my_list = [20,30,40,50]
# For swapping values, we will first use a temporary value.
temp = my_list[0]
my_list[0] = my_list[3]
my_list[3] = temp
print(my_list)

```

```python
#Swapping values, without using temporary values
my_list = [5,6,9,2]
print(my_list[0])
print(my_list[3])
my_list[0],my_list[3] = my_list[3],my_list[0]
print(my_list)
```

```python
#Create a dictionary, do all operations like addition of elements, remove elements, copy item & create nested dictionaries too.
```

```python
Student_info= {"Name": "Harshita", "Age": 18, "Gender":"Female"}
```

```python
#addition of element in a dictionary
Student_info["Course"]= "Artificial Intelligence"
print(Student_info)
```

```python
#Update can be use to add new values or update the existing value
Student_info.update({"Passing year":2025}) #for adding new values
print(Student_info)
```

```python
Student_info.update({"Course":"Datascience"}) #for updating preexisting values
print(Student_info)
```

```python
#Removing of elements
#popitem method
Student_info.popitem()      #for removing the last element
print(Student_info)
```

```python
#pop
Student_info.pop("Age")
print(Student_info)
```

```python
#del is used to remove item with specific key name
del Student_info ['Gender']
print(Student_info)
```

```python
#to delete whole dictionary
del Student_info
```

```python
#to empty the dictionary we use (.clear)
Student_info.clear()
print(Student_info)
```

```python
# To copy a dictionary  (.copy) is used
Id_info = Student_info.copy()
print(Id_info)
```
