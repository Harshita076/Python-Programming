```python
#Union,it joins two sets.
set1 = {1,2,3}
set2 = {4,5,7}
set3 = set2.union(set1)
print(set3)
```

```python
# | (pipeline method)
set1 = {5,8,9,0}
set2 = {8,7,0,9}
set3 = set2|(set1)
print(set3)
```

```python
# intersection method
set1 = {5,8,9,0}
set2 = {3,2,0,9}
set3 = set1.intersection(set2)
print(set3)
```

```python
# & operator
set1 = {5,8,9,0}
set2 = {3,2,0,9}
set3 = set1 & set2
print(set3)
```

```python
set1 = {5,8,9,0}
set2 = {3,2,0,9}
set3 = set2.difference(set1)
print(set3)
```
