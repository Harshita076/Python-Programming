```python
#function to swap two values
def swap(a,b):
    temp=a
    a=b
    b=temp
    print(a)
    print(b)
    
```

```python
swap(3,7)
```

```python
def swap(a,b):
    a,b=b,a
    print(a)
    print(b)
```

```python
swap(5,8)
```
