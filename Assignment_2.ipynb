```python
#Write a Python program that prints all even numbers between 1 and 50 using a for loop, skipping multiples of
#10 with the continue statement. Include an option to exit the loop early if a number greater than 40 is encountered
#using the break statement.
for i in range(1, 51):
    if i % 2 == 0:     
        if i % 10 == 0:  #is being used skipping multiple of 10
            continue
        if i > 40:       # for exiting loop early if number is greater than 40
            break
        print(i)
```

```python
#Write a program to check for Palindrome, Armstrong & Fibonacci Series.

#For Palindrome: A palindrome is a word, number, phrase, or sequence that reads the same backwards as forward.
x = input("Enter a word/number: ")
x=x.lower()
if x == x[::-1]:  #it is for reversing the string.
    print("Its a Palindrome.")
else:
    print("Not a Palindrome.")
```

```python
#  For Armstrong: An Armstrong number is a number that is equal to the sum of its own digits, each raised to the power of the number of digits.
x = int(input("Enter a number: "))
sum = 0
temp = x
n = len(str(x))

while temp > 0: 
    digit = temp % 10
    sum += digit ** n
    temp //= 10  

if sum == x:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
## Fibonacci: The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones.
n = int(input("Enter number of terms: "))

a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Create a Python module that contains a function to calculate the factorial of a number using recursion

#factorial: The factorial of a non-negative integer n is the product of all positive integers less than or equal to n
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:         #Multiply n by the factorial of (n-1)        
        return n * factorial(n - 1)    
        
num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```
