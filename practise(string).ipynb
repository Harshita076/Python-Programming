```python
#Nested for loop
```

```python
adj = ["red","yellow","bitter"]
fruit = ["apple","banana","mango"]
for x in adj:
    for y in fruit:
        print(x,y)
```

```python
adj = ["red","pink","yellow"]
thing = ["table","chair","pen"]
for x in adj:
    for y in thing:
        print(x,y)
```
