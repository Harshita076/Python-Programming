```python
#Scatter PLot
df2 = sns.load_dataset("titanic")
sns.scatterplot(x="age", y="fare", hue="survived", data=df2)
plt.show()
```

```python
#Line Chart
sns.lineplot(x="age", y="fare", data=df2)
plt.show()
```

```python
#Relational PLot(relplot)
sns.relplot(x="age", y="fare", hue="sex", col="survived", data=df2)
plt.show()
```

```python
#Histogram
sns.histplot(df2["age"], bins=50, kde=False)
plt.show()
```

```python
#KDE PLot (kernel density estimate)
sns.kdeplot(df2["age"], fill=True)
plt.show()
```

```python
#Distribution Plot (Histogram + KDE)
sns.histplot(df2["fare"], bins=50, kde=True)
plt.show()
```

```python
#Bar Chart
sns.barplot(x="sex", y="survived", data=df2)
plt.show()
```

```python
#Boxplot
sns.boxplot(x="Pclass", y="Age", data=df2)
```

```python
#Violin Plot
sns.violinplot(x="Sex", y="Age", hue="Survived", data=df2, split=True)
```

```python
#Heatmap    #Here Red colour represents- positive corerelation whereas blue colour shows negative corerelation 
corr = df2.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Pair plot
sns.pairplot(df2, hue="Survived")
plt.show()
```

```python
#Countbar 
sns.countplot(x="Pclass", hue="Survived", data=df2)
plt.show()
```
